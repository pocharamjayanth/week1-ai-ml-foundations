```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the data you just uploaded
df = pd.read_csv("../data/student_scores.csv")

# 2. Inspect the data
print(df.head())
print(df.info())

# 3. Create the required visualization
plt.figure(figsize=(8,5))
sns.scatterplot(x='Hours_Studied', y='Scores', data=df)
plt.title('Study Hours vs Student Scores')
plt.show()
```

```python
# Check for missing values
print("\nMissing values check:")
print(df.isnull().sum())

# Rename columns to lowercase for consistency
df.columns = ["hours_studied", "score"]

# Save the cleaned version
df.to_csv("../data/student_scores_cleaned.csv", index=False)
print("\nCleaned data saved as 'student_scores_cleaned.csv'")
```

```python
# Create a scatter plot to show the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='hours_studied', y='score', data=df, color='blue', s=100)
plt.title('Relationship Between Study Hours and Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Calculate Correlation
correlation = df['hours_studied'].corr(df['score'])
print(f"The correlation between study hours and scores is: {correlation:.2f}")
```
