Data Exploration & Visualization Code

```python
# Cell 1: Setup and Data Creation
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

# Since we are in Colab, we can create the data directly to ensure it works
data = """Hours_Studied,Scores
2.5,21
5.1,47
3.2,27
8.5,75
3.5,30
1.5,20
9.2,88
5.5,60
8.3,81
2.7,25
7.7,85
5.9,62
4.5,41
3.3,42
1.1,17
8.9,95
2.5,30
1.9,24
6.1,67
7.4,69
2.7,30
4.8,54
3.8,35
6.9,76
7.8,86"""

# Load the dataset
df = pd.read_csv(io.StringIO(data))
print("Dataset Loaded Successfully!")
print(df.head()) # Shows first 5 rows

# Cell 2: Data Cleaning (Step 5 of your guide)
# Check for missing values
print("\nMissing values check:")
print(df.isnull().sum())

# Rename columns to lowercase for consistency
df.columns = ["hours_studied", "score"]

# Save the cleaned version locally in Colab
df.to_csv("student_scores_cleaned.csv", index=False)
print("\nCleaned data saved as 'student_scores_cleaned.csv'")

# Cell 3: Visualization (Step 6 of your guide)
# 1. Distribution of Study Hours
plt.figure(figsize=(8, 4))
plt.hist(df["hours_studied"], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Hours Studied")
plt.ylabel("Frequency")
plt.title("Distribution of Study Hours")
plt.show()

# 2. Study Hours vs Score Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="hours_studied", y="score", data=df, color='red', s=100)
plt.title("Study Hours vs Student Performance")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Cell 4: Actionable Insights (Step 7 of your guide)
correlation = df['hours_studied'].corr(df['score'])
print(f"\nStatistical Insight: The correlation between study time and score is {correlation:.2f}")
print("Conclusion: There is a strong positive correlation between study hours and academic performance.")
```
